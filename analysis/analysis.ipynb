#####  Notebook to create quick analysis on model size and training time

```python
import os
import matplotlib.pyplot as plt

# Data
models = ['Swin-Tiny', 'ViT', 'BEiT', 'DEiT']
training_times = [13481.891012430191, 23981.2263379097, 20445.953822374344, 11973.28246474266]  # Total training time for each model (in hours)
epochs = [35, 30, 25, 15]  # Number of epochs for each model

# Calculate training time per epoch
time_per_epoch = [time / epoch for time, epoch in zip(training_times, epochs)]

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(models, time_per_epoch, color='teal', width=0.5)

# Add labels and title
plt.xlabel('Models', fontsize=13)
plt.ylabel('Training Time per Epoch (hours)', fontsize=13)
plt.title('Comparison of Models by Training Time per Epoch', fontsize=13)

# Show plot
plt.show()
```

```python
# File paths of the model checkpoint files (.pth files)
model_paths = ['models/trained_beit_model.pth', 'models/trained_vit_model.pth', 'models/trained_deit_model.pth', 'models/trained_swin_tiny.pth']

# Function to get the size of a file in MB
def get_file_size(file_path):
    return os.path.getsize(file_path) / (1024 * 1024)  # Convert bytes to MB

# Get sizes of all model files
model_sizes = [get_file_size(model_path) for model_path in model_paths]

# Plotting the sizes
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(model_paths)+1), model_sizes, marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Models', fontsize=13)
plt.ylabel('Model Size (MB)', fontsize=13)
plt.title('Comparison of Model Sizes', fontsize=13)

# Add model labels
plt.xticks(range(1, len(model_paths)+1), ['BEiT', 'ViT', 'DEiT', 'Swin-Tiny'])

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

# Show plot
plt.grid(True)
plt.show()
```
